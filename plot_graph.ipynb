```python
import networkx as nx 
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_node("H")



```

```python
G = nx.erdos_renyi_graph(10, 0.5)
pos = nx.spring_layout(G)
```

```python

theta =  np.pi / 2
V = np.array([[np.sin(theta)],[np.cos(theta)]])
heights = {}
for key, val in pos.items():
    heights[key] = val @ V 

hbins = np.linspace(np.stack(heights.values()).min()-.01, np.stack(heights.values()).max()+.01,4,endpoint=True)

```

```python
for level in hbins:
    color_map = []
    for node, val in heights.items():
        if val < level :
            color_map.append('black')
        else: 
            color_map.append('lightgray')      
    nx.draw(G, pos = pos, node_color=color_map)
    plt.show()
```
